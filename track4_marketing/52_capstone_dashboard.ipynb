## §2 · Setup & Imports

```python
# ── Reproducibility ──────────────────────────────────────────────
import numpy as np
import pandas as pd
import warnings

np.random.seed(42)
warnings.filterwarnings("ignore")

# ── Visualization ────────────────────────────────────────────────
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

# ── Display settings ─────────────────────────────────────────────
pd.set_option("display.max_columns", 30)
pd.set_option("display.float_format", lambda x: f"{x:,.2f}")

print("✓ All imports loaded")
```

---
## §3 · Synthetic Business Data

```python
# Generate synthetic business data
N_CUSTOMERS = 10000
N_TRANSACTIONS = 100000

# Customer data
customers = pd.DataFrame({
    'customer_id': range(N_CUSTOMERS),
    'age': np.random.normal(35, 10, N_CUSTOMERS).clip(18, 70).astype(int),
    'tenure_months': np.random.exponential(24, N_CUSTOMERS).astype(int),
    'monthly_spend': np.random.exponential(100, N_CUSTOMERS),
    'churn_probability': np.random.beta(2, 8, N_CUSTOMERS),
    'clv': np.random.exponential(500, N_CUSTOMERS),
})

# Transaction data
transactions = pd.DataFrame({
    'transaction_id': range(N_TRANSACTIONS),
    'customer_id': np.random.randint(0, N_CUSTOMERS, N_TRANSACTIONS),
    'amount': np.random.exponential(50, N_TRANSACTIONS),
    'is_fraud': np.random.binomial(1, 0.02, N_TRANSACTIONS),
    'date': pd.date_range('2024-01-01', periods=N_TRANSACTIONS, freq='1h'),
})

# Marketing data
marketing = pd.DataFrame({
    'channel': ['TV', 'Social', 'Search', 'Email'],
    'spend': [500000, 200000, 300000, 50000],
    'conversions': [2000, 1500, 3000, 1000],
    'revenue': [1000000, 600000, 900000, 300000],
})

print(f"✓ Generated synthetic business data")
print(f"  Customers: {N_CUSTOMERS:,}")
print(f"  Transactions: {N_TRANSACTIONS:,}")
print(f"  Marketing channels: {len(marketing)}")
```

---
## §4 · Fraud Detection Dashboard

```python
# Fraud metrics
fraud_rate = transactions['is_fraud'].mean() * 100
fraud_amount = transactions[transactions['is_fraud'] == 1]['amount'].sum()
total_amount = transactions['amount'].sum()

print("=" * 70)
print("FRAUD DETECTION DASHBOARD")
print("=" * 70)
print(f"\nFraud Rate: {fraud_rate:.2f}%")
print(f"Fraud Amount: ${fraud_amount:,.2f}")
print(f"Total Amount: ${total_amount:,.2f}")
print(f"Fraud Loss Rate: {fraud_amount/total_amount*100:.2f}%")

# Fraud by hour
transactions['hour'] = transactions['date'].dt.hour
fraud_by_hour = transactions.groupby('hour')['is_fraud'].mean() * 100

fig = go.Figure()
fig.add_trace(go.Bar(
    x=fraud_by_hour.index,
    y=fraud_by_hour.values,
    name='Fraud Rate by Hour'
))
fig.update_layout(title='Fraud Rate by Hour of Day', height=300)
fig.show()
```

---
## §5 · Churn & CLV Dashboard

```python
# Churn & CLV metrics
avg_churn = customers['churn_probability'].mean() * 100
avg_clv = customers['clv'].mean()
total_clv = customers['clv'].sum()

print("=" * 70)
print("CHURN & CLV DASHBOARD")
print("=" * 70)
print(f"\nAverage Churn Probability: {avg_churn:.1f}%")
print(f"Average CLV: ${avg_clv:,.2f}")
print(f"Total CLV: ${total_clv:,.2f}")

# CLV distribution
fig = go.Figure()
fig.add_trace(go.Histogram(
    x=customers['clv'],
    nbinsx=50,
    name='CLV Distribution'
))
fig.update_layout(title='Customer Lifetime Value Distribution', height=300)
fig.show()
```

---
## §6 · Marketing Attribution Dashboard

```python
# Marketing metrics
marketing['roi'] = (marketing['revenue'] - marketing['spend']) / marketing['spend'] * 100
marketing['cac'] = marketing['spend'] / marketing['conversions']

print("=" * 70)
print("MARKETING ATTRIBUTION DASHBOARD")
print("=" * 70)
print(f"\nChannel Performance:")
print(marketing[['channel', 'spend', 'revenue', 'roi', 'cac']].to_string(index=False))

# ROI comparison
fig = go.Figure()
fig.add_trace(go.Bar(
    x=marketing['channel'],
    y=marketing['roi'],
    name='ROI (%)'
))
fig.update_layout(title='Marketing ROI by Channel', height=300)
fig.show()
```
